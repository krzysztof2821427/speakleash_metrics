#### Adding ratios

```python
nlp = spacy.load("pl_core_news_sm")

pos_dict = {
    'prop_nouns': 'PROPN',
    'adjectives': 'ADJ',
    'adverbs': 'ADV',
    'pronouns': 'PRON',
    'determiners': 'DET',
    'conjuctions': 'CONJ',
    'auxverbs': 'AUX', 
    'cardinal_numbers': 'NUM'    
}


def add_ratio(col1, col2=df['words']):
    return round(col1.div(col2), 2)


def add_pos_counts(df, pos_dict):    
    for key, value in pos_dict.items():        
        df[key] = df['pos_tag'].apply(lambda x: len([1 for tag in x if tag.endswith(value)]))        
        new_col = f'{key}_ratio'        
        df[new_col] = add_ratio(df[key])
    return df
```

```python
df['pos_tag'] =  df['text'].apply(lambda x: [token.pos_ for token in nlp(x)])

    

with open("df_pos_added.pkl","wb") as f:
    pickle.dump(df, f)
```

```python
df['avg_sent_len'] = add_ratio(df['words'], df['sentences'])
df['verbs_ratio'] = add_ratio(df['verbs'])
df['nouns_ratio'] = add_ratio(df['nouns'])
df['punctuations_ratio'] = add_ratio(df['punctuations'])
df['symbols_ratio'] = add_ratio(df['symbols'])
df['stopwords_ratio'] = add_ratio(df['stopwords'])

df = add_pos_counts(df, pos_dict)
```

```python
pd.set_option('max_colwidth', None)
df.head()
```

```python
df['readability_ind'] = df['text'].apply(textstat.automated_readability_index)
df['long_words'] = df['text'].apply(textstat.long_word_count)
df['long_words_ratio'] = add_ratio(df['long_words'])
```

```python
df['gunning_fog'] = df['text'].apply(textstat.gunning_fog)
```
