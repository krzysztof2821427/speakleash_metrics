```python
import Levenshtein as lev
import numpy as np
import pandas as pd
import pickle
import plotly.express as px
import plotly.graph_objects as go

#!pip install levenshtein
```

```python
with open("df_feat_added_full.pkl","rb") as f:
    df = pickle.load(f)
```

```python
df.columns
```

```python
compare_df = df[['name', 'text', 'length', 'sentences', 'words', 'prop_nouns']][df[['length', 'sentences', 'words', 'prop_nouns']].duplicated(keep=False) == True]
```

```python
compare_df.sort_values(by='sentences')
```

```python
def list_strings(df,col):
    strdict = {}
    for i in set(df[col].values):
        strdict[i] = df['text'][df[col] == i].values.tolist()
    return strdict

def get_lev_median(strdict):
    median_dict = dict()
    for key, value in strdict.items():
         median_dict[key] = lev.quickmedian(value)
    return median_dict

def add_to_df(x):
    for key, value in median_dict.items():
        if x == key:
            return value
```

```python
strdict = list_strings(compare_df, 'sentences')
median_dict = get_lev_median(strdict)
compare_df['median_str'] = compare_df['sentences'].apply(add_to_df)
with open("compare_df","wb") as f:
    pickle.dump(compare_df, f)
```

```python
compare_df['lev_zip'] = list(zip(compare_df['median_str'].values, compare_df['text'].values))
```

```python
compare_df['lev_ratio'] = compare_df['lev_zip'].apply(lambda x: lev.ratio(x[0], x[1]))
```

```python
compare_df['lev_ratio'].value_counts()
```

```python
fig = px.scatter_polar(compare_df, r=compare_df['lev_ratio'], theta=compare_df['prop_nouns'], color=compare_df['sentences'], color_continuous_scale=px.colors.cyclical.HSV, template="plotly_dark")
fig.show()
```

```python
fig = px.histogram(compare_df['lev_ratio'])
fig.show()
```
